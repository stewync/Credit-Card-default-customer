```python
file = file.rename(columns = {'default payment next month': 'def_pay_next'})
```

```python
#Probability of Default payment for next month

P_def_pay = file['def_pay_next'].sum()/ len(file['def_pay_next'])
P_def_pay
```

```python
# counting #Men/women in  sex column
file['SEX'].value_counts()
```

```python
# As we have 0 undocumented in sex we convert it into 3(other)
file['MARRIAGE'].value_counts()
```

```python
#Replace 0 to 3 in marriage
file['MARRIAGE'] = file['MARRIAGE'].replace(0,3)
file['MARRIAGE'].value_counts()
```

```python
# Replace 0,5 & 6 in education to 4(others)
file['EDUCATION'] = file['EDUCATION'].replace([0,5,6],4)
file['EDUCATION'].value_counts()
```

```python
#Groupby for Sex and default payment 
sex_def_pay = file.groupby(['SEX', 'def_pay_next']).size()
sex_def_pay
```

```python
#Groupby for Education and default payment
education_def_pay = file.groupby(['EDUCATION','def_pay_next']).size()
education_def_pay
```
