###  Education has values which range from 0 to 6 where (0,5,6 are undocumented )

```python
file['EDUCATION'].value_counts()
```

#### Replace all undocumented values in Education (0,5,6) to 4 -> 'others' 

1 -> Graduate School

2 -> University

3 -> High School

4 -> Others

```python
file['EDUCATION'] = file['EDUCATION'].replace([0,5,6],4)
education_count = file['EDUCATION'].value_counts()
print(education_count)
education_count.plot.pie(labels =['University','Graduate School','High School','Others'],autopct='%.2f')
```

#### Marriage | Replace undocumented value 0 to 3 (others)
1 -> Married

2 -> Single

3 -> Others

```python
file['MARRIAGE'] = file['MARRIAGE'].replace(0,3)
marriage_count = file['MARRIAGE'].value_counts()
print(marriage_count)
marriage_count.plot.pie(labels = ['Single','Married','Others'],autopct='%.2f')
```

### Sex

```python
sex_count = file['SEX'].value_counts()
print(sex_count)
sex_count.plot.pie(labels = ['Female','Male'], autopct = '%.2f')
```

### Categorical variable [Sex, Marriage,Education] 

```python
categorical_vars = ['EDUCATION','SEX','MARRIAGE']
# Changing the type of categorical variable to string
file[categorical_vars] = file[categorical_vars].astype(str)
file = pd.get_dummies(file, columns=categorical_vars)

print(file.head())
```

### History of past payments ['PAY_1' - 'PAY_6']

#### From the description of history of past payment minimum value is -2 which I assume payment made in advance 2 months before.


-1 : Pay one month ahead || -2 : Pay two months ahead || 0 : Pay duly

1 : Payment delay for 1 month ..... 8: Payment delay for 8 month

PAY_1 : Repayment Status in September

PAY_2 : Repayment Status in August

PAY_3 : Repayment Status in July
